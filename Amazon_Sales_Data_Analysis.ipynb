### H. Handling nilai Rupe ke Rupiah

Diskon dan harga aktual dalam Rupee. Konversikan Rupee India ke dalam dolar AS. Nilai tukar Rupee ke dolar adalah 0,012.

```python
# Nilai tukar dari Rupee India ke Rupiah Indonesia
# Nilai ini mungkin berubah-ubah, jadi pastikan untuk memeriksa nilai tukar terkini
exchange_rate = 0.012

# Konversi harga dari Rupee India ke Rupiah Indonesia
amazon['discounted_price'] = amazon['discounted_price'].apply(lambda x: x * exchange_rate)
amazon['actual_price'] = amazon['actual_price'].apply(lambda x: x * exchange_rate)

# Bulatkan harga ke 2 angka desimal
amazon['discounted_price'] = round(amazon['discounted_price'], 2)
amazon['actual_price'] = round(amazon['actual_price'], 2)
```

# 5.**Handling Missing Value**

```python
# Gunakan perintah .isna() dan .sum() untuk melihat berapa banyak nilai null yang ada di setiap kolom.
amazon.isna().sum()
```

Dari informasi yang didapat, **terdapat missing value** pada kolom `rating_count` sebanyak 2 item

## A. Rating_count

```python
# Nilai 2 null terletak di kolom rating_count.
# Gunakan perintah .index untuk menemukan nomor indeks baris 1 atau lebih nilai null.
# Gunakan perintah .iloc untuk menampilkan baris berdasarkan nomor indeks.

amazon.iloc[amazon[(amazon.isna().sum(axis = 1) >= 1)].index]
```

```python
# Ganti nilai null dengan rata-rata jumlah_rating.
# Gunakan perintah .mean() untuk menghitung rata-rata rating_count.
# Gunakan perintah round() untuk membulatkan rata-rata ke bilangan bulat terdekat.
# Gunakan perintah .fillna() untuk mengganti nilai null dengan nilai rata-rata yang dibulatkan.

amazon['rating_count'] = amazon['rating_count'].fillna(round(amazon["rating_count"].mean()))
```

# 9.Membuat kolom yang mengurutkan produk berdasarkan skor penilaian.
Kolom peringkat akan berisi peringkat berdasarkan peringkat produk. Setiap peringkat akan mewakili rentang peringkat (1-2 bintang = Buruk).

```python
# Buat perulangan for dan pernyataan if else untuk mengkategorikan skala peringkat.
# Gunakan perintah .append() untuk menambahkan kategori peringkat ke daftar peringkat.

ranking = []

for score in amazon['rating']:
    if score <= .9 : ranking.append('Very Poor')
    elif score <= 1.9 : ranking.append('Poor')
    elif score <= 2.9 : ranking.append('Average')
    elif score <= 3.9 : ranking.append('Good')
    elif score <= 4.9 : ranking.append('Very Good')
    elif score == 5.0 : ranking.append('Excellent')

# Tambahkan daftar peringkat sebagai kolom peringkat dalam bingkai data amazon.
# Gunakan perintah .astype() untuk mengonversi nilai dalam kolom peringkat menjadi tipe data kategori.

amazon['ranking'] = ranking

amazon['ranking'] = amazon['ranking'].astype('category')
```

```python
# Hitung koefisien korelasi Pearson (default di Pandas)
correlation_matrix = amazon.corr()

# Cetak matriks korelasi
print(correlation_matrix)

# Buat heatmap untuk memvisualisasikan korelasinya
sns.heatmap(correlation_matrix, annot=False, cmap="coolwarm")
plt.title("Correlation Matrix (Pearson)")
plt.show()

# Hitung koefisien korelasi Spearman (untuk hubungan non-linier)
spearman_correlation_matrix = amazon.corr(method="spearman")

# Cetak matriks korelasi Spearman
print(spearman_correlation_matrix)

# Buat heatmap untuk memvisualisasikan korelasi Spearman
sns.heatmap(spearman_correlation_matrix, annot=False, cmap="coolwarm")
plt.title("Correlation Matrix (Spearman)")
plt.show()
```
